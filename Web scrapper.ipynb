### Buiding a web scrapper

```python
import requests
#import numpy as np
from bs4 import BeautifulSoup as bs
import pandas as pd
```

### URL to scrape from

```python
url = url = 'https://en.wikipedia.org/wiki/List_of_universities_in_Nigeria'
```

### Requesting for the data

```python
web_data = requests.get(url)
```

```python
scrapped_data = web_data.content
```

```python
scrapped_data
```

### Using Beautiful soup to make data clean 

```python
soup = bs(scrapped_data, 'lxml')
```

```python
soup
```

```python
list_of_rows = []
for row in soup.findAll('tr'):
    list_of_cells = []
    for cell in row.findAll('td'):
        #text = cell.text.encode('utf-8')
        text = cell.text
        text = text.replace('\n', ' ')
        #print(text)
        list_of_cells.append(text)
    list_of_rows.append(list_of_cells)
```

```python
list_of_cells
```

```python
list_of_rows
```

### Putting scrapped data in a table using Pandas

```python
wiki_data = pd.DataFrame(list_of_rows[1:], columns= ['University Name', 'State', 'Abbrevation', 'Location', 'Status', 'Year found'])
```

```python
wiki_data.dropna()
```

### Simple data Exloration

### Export scrapped data to a CSV

```python
wiki_data.to_csv("University_Data.csv")
```
